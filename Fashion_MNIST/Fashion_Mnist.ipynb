```
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

# ***Fashion Mnist***

```
data=keras.datasets.fashion_mnist
```

```
(train_image,train_label), (test_image,test_label)= data.load_data() #() are necessary as 2 tuples are o/p not 4 values
```

```
(train_image.size)# 784*60k(number of images/labels)
```

```
(train_label.size)# 60k(number of labels)
```

```
# pixels size is 28*28=784.
```

```
plt.imshow(train_image[10],cmap=plt.cm.binary)
```

```
ctrain_image=train_image/255.0
ctest_image=test_image/255.0
```

```
mod=keras.Sequential([
     keras.layers.Flatten(input_shape=(28,28)),
     keras.layers.Dense(128,activation='relu'),
     keras.layers.Dense(10,activation='softmax')                    
])
```

# ***Q) tf.keras.models.sequential vs tf.keras.sequential?***
# ***Q) Can all layers be added in one such thing or different obj reqd?***

```
mod.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```
mod.fit(ctrain_image,train_label,epochs=10)
```

```
t_loss,t_acc=mod.evaluate(ctest_image,test_label)
```

```
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```
prediction=mod.predict(ctest_image)
```

```
for i in range(15):
  plt.imshow(test_image[i],cmap=plt.cm.binary)
  plt.xlabel("Actual= "+class_names[test_label[i]])
  plt.title("Prediction= "+class_names[np.argmax(prediction[i])])
  plt.show()
```
