# **Data Prepeocessing**
## Train data preprocessing

```
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)

training_set = train_datagen.flow_from_directory(
        train_path,
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary')
```

## Test Data Preprocessing

```
test_datagen=ImageDataGenerator(rescale=1./255) #Only scaling, no other transformations on Test data

test_set = test_datagen.flow_from_directory(
        test_path,
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary')
```

# CNN Layers

## 1.Initialising Cnn layer

```
cnn=tf.keras.models.Sequential()
```

## 2. Cnn

```
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3,activation='relu', input_shape=[64,64,3])) #I/p shape:- 64 X 64 was mentioed in preprocessing and 3 due to RGB Layers
```

## 3.Pooling

```
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

## Adding second convolution layers

```
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu')) #I/p shape parameter removed. Only reqd for 1st layer
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

## 4.Flatten

```
cnn.add(tf.keras.layers.Flatten())
```

## Full Connection

```
cnn.add(tf.keras.layers.Dense(128,activation='relu'))
```

## Output Layer

```
cnn.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

## Compiling

```
cnn.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

## Train Cnn and validation on Test set

```
cnn.fit(x=training_set, validation_data=test_set, epochs=25)
```
